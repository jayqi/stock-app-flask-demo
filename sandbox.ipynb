```python
import requests
import pandas as pd
import datetime
import dateutil
import re


from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
tickersymbol = 'AAPL'
url = 'https://www.quandl.com/api/v3/datasets/WIKI/' + tickersymbol + '.json'
r = requests.get(url)
r.json()
```

```python
request_df = pd.DataFrame(r.json())
request_df
```

```python
df = request_df['dataset'].apply(pd.Series) 
df.head(5)
```

```python
data = df.ix['data',:].apply(pd.Series) 
data.head(5)
```

```python
data.columns = df.ix['column_names',0:12]
data.head(5)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.dtypes
data.head(5)
```

```python
output_notebook()
```

```python
today = datetime.date.today()
today
```

```python
lastmonth = today - dateutil.relativedelta.relativedelta(months=1)
lastmonth
```

```python
b = data['Date']>lastmonth
b.head(25)
```

```python
data_lastmonth = data[b]
data_lastmonth
```

```python
p = figure(title=tickersymbol,plot_width=600,plot_height=400,x_axis_type="datetime",x_range=(lastmonth,today))
p.line(data_lastmonth['Date'],data_lastmonth['Open'])

p.xgrid.minor_grid_line_color = 'navy'
```

```python
show(p)
```
